```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_xy(x, y, title):
    fig, ax = plt.subplots(1,figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    ax.set_xlabel('Eixo x')
    ax.set_ylabel('Eixo y')
    plt.show()
```

```python
## 'f x = case (x ≤ -1): x + 1; case (-1 < x ≤ 2): -x² + 2; case (x > 2): x
x = np.arange(-15, -1, 0.1)
y = x + 1
plt.plot(x , y, label = 'case (x ≤ -1): x + 1')
x2 = np.arange(-1, 2, 0.1)
y2 = 2 - np.power(x2, 2)
plt.plot(x2 , y2, label = 'case (-1 < x ≤ 2): -x² + 2')
x3 = np.arange(2, 15, 0.1)
plt.plot(x3 , x3, label = 'case (-1 < x > 2): x')
plt.title('f x = case (x ≤ -1): x + 1; case (-1 < x ≤ 2): -x² + 2; case (x > 2): x')
```
