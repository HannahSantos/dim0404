```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
```

```python
def plot_xy(x,y, title):
    fig, ax = plt.subplots(1, figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    plt.show()
```

```python
x = np.arange(-50,50,0.1)
x
```

```python
## f x = x**(1/3) + x**3 + 7
y = np.cbrt(x) + x**3 + 7
plot_xy(x, y, 'f x = x**(1/3) + x**3 + 7')
```

```python
## f x = |x|
y = np.abs(x)
plot_xy(x, y, 'f x = |x|')
```

```python
## f x = sin x
y = np.sin(x)
plot_xy(x, y, 'f x = sin x')
```

```python
x_pos = np.arange(0.0001,50,0.1)
x_neg = np.arange(-50, -1,0.1)
```

```python
## f x = ln x
y = np.log(x_pos)
plot_xy(x_pos, y, 'f x = ln x')
```

```python
## f x = e**x
y = np.exp(x)
plot_xy(x, y, 'f x = e**x')
```
