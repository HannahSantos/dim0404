```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_xy(x, y, title):
    fig, ax = plt.subplots(1,figsize=(10,5))
    ax.plot(x,y)
    ax.set_title(title)
    ax.set_xlabel('Eixo x')
    ax.set_ylabel('Eixo y')
    plt.show()
```

```python
## 'f x = case (x ≤ -1): x + 1; case (-1 < x ≤ 2): -x² + 2; case (x > 2): x

x_balls1 = [-1, 2]
y_balls1 = [0, -2]

x = np.arange(-10, -1, 0.01)
y = x + 1
plt.plot(x , y, label = 'x + 1       , x ≤ -1')
x2 = np.arange(-1, 2, 0.01)
y2 = 2 - np.power(x2, 2)
plt.plot(x2 , y2, label = '-x² + 2     , -1 < x ≤ 2')
x3 = np.arange(2, 10, 0.01)
plt.plot(x3 , x3, label = 'x              , x > 2')

plt.scatter(x_balls1, y_balls1, label= "belongs", color= "blue", 
            marker= "o", s=12)
x_balls2 = [-1, 2]
y_balls2 = [1, 2]
plt.scatter(x_balls2, y_balls2, label= "doesn't belong", color= "gray", 
            marker= "o", s=12)

plt.legend()
plt.title('f x')
```
